```python
from music21 import *
import random

s1 = stream.Stream();
s1.clear()
s1.insert(0, metadata.Metadata())
s1.metadata.title = '  '
s1.metadata.composer = ''
s1.insert(0, meter.TimeSignature('12/8'))
s1.keySignature= key.KeySignature(-3)

for i in range(100):
    note_or_rest = random.randrange(1, 10)
    print(note_or_rest)
    if note_or_rest>6:
        n1 = note.Rest(55)
        random_length = random.randrange(1, 4) * 0.25
        n1.quarterLength=random_length
        s1.append(n1)
        
        print("")   
    else:
        randomnote =random.randrange(1, 6)
        if randomnote == 1:
            n1 = note.Note(55)
            n1.quarterLength=0.5
            s1.append(n1)
        if randomnote == 2:
            n1 = note.Note(57)
            n1.quarterLength=0.5
            s1.append(n1)
        if randomnote == 3:
            n1 = note.Note(60)
            n1.quarterLength=0.5
            s1.append(n1)
        if randomnote == 4:
            n1 = note.Note(62)
            n1.quarterLength=0.5
            s1.append(n1)
        if randomnote == 5:
            n1 = note.Note(64)
            n1.quarterLength=0.125
            s1.append(n1)
            n1 = note.Note(62)
            n1.quarterLength=0.375
            s1.append(n1)
        print("")   
```
